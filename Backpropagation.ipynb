```python
import numpy as np
```

```python
x=10
y=20
bais = 0.001
weight = 10
eta = 0.01
epoch =100
```

```python
for i in range(0, epoch):
    o = weight * x + bais
    dw = (o-y)*x
    db = (o-y)*1
    
    weight = weight-eta*dw
    bais = bais - eta*db
    c = (o-y)**2
    print(" o = "+str(o)+ " cost = "+str(c))
```
